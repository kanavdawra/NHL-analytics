```python
import time as time

from selenium import webdriver
from selenium.webdriver.support.select import Select
import pandas as pd
```

```python
def custom_split(str):
    data = str.split(' ')
    final = []
    index = len(data)-1
    while index>=0:

        element = data[index]
        if data[index-1] == 'OT' or data[index-1] == 'SO':
            element = data[index-1]+' '+data[index]
            index -= 1
        try:
            if data[index + 1] == 'Home' or data[index + 1] == 'Away':
                team = ''
                for _ in range(2, index+1):
                    team = team + data[_] + ' '

                final.insert(0,team.strip())
                final.insert(0,data[1])
                final.insert(0,data[0])
                break
        except:
            final.insert(0,element)
            index -= 1
            continue


        final.insert(0,element)
        index -= 1

    return final
```

```python
driver = webdriver.Chrome('E:/Projects/Webscrapes/chromedriver')
driver.get('https://www.quanthockey.com/nhl/team-game-logs/anaheim-ducks-2022-23-nhl-game-log.html')



# season_type = driver.find_element_by_id('st-filter')
# season_type_select = Select(season_type)

teams = []
df = pd.DataFrame(columns=['Rank', 'Date', 'Challenger','Opponent', 'Location',
                           'Result', 'GF', 'GA', 'GD', 'PDO', 'PDO-A', 'SF',
                           'SA', 'SD', 'SH%', 'SH%-A', 'FOW', 'FOL', 'FOD',
                           'FO%', 'SV%', 'SV%-A', 'HITS', 'HITS-A', 'HITS-D',
                           'BS', 'BS-A', 'BS-D'])

for _ in range(0,7):

    time.sleep(2)
    season = driver.find_element_by_id('season-filter')
    season_select = Select(season)
    season_select.select_by_index(_)
    index = 0
    while True:


            time.sleep(2)
            team = driver.find_element_by_id('team-filter')
            team_select = Select(team)
            try:
                team_select.select_by_index(index)
                index += 1
                time.sleep(2)
            except:
                break
            trs = driver.find_elements_by_xpath('//*[@id="statistics"]/tbody/tr')
            print(driver.find_element_by_xpath(f'//*[@id="team-filter"]/optgroup/option[{index}]').text)
            for _ in range(0,len(trs)):
                data = custom_split(trs[_].text)
                data.insert(2, driver.find_element_by_xpath(f'//*[@id="team-filter"]/optgroup/option[{index}]').text)
                print(_,len(data))
                df.loc[df.shape[0] - 1] = data

            print(df.shape)
    index = 0
```
